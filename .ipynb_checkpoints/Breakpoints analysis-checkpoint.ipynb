```python
import json
import matplotlib.pyplot as plt

from changes import jumps,plot
from stasts import filter_outliers
from read_file import load_data,read_file_original
```

```python
source = 'others/wosPlosOne2016_citations.json'
text = open(source,'r').read()
json_data = json.loads(text)
```

```python
data = load_data('data/plos_one_total_breakpoints_k4_original1_data.txt')
data = filter_outliers(data)
# top 10, descending
rankings = jumps(data,10,True)
```

```python
for n_intervals,ranking in rankings.items():
    print("Number of intervals %d" % n_intervals)
    for sample in ranking:
        idx = sample[0]
        print(idx)
        sample_info = json_data[idx]
        print(sample_info['DI'])
        X = sample[3]
        Y = sample[4]
        Y_pred = sample[-1]
        print(X[0],Y[0])
        print(Y_pred)
        plt.scatter(X[1:],Y[1:],color='blue')
        plt.scatter(X[1:],Y_pred,color='red')
        plt.show()
    print()
```
